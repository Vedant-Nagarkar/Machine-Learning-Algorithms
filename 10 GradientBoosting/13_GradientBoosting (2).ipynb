```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.metrics import accuracy_score
from sklearn.ensemble import GradientBoostingClassifier
```

```python
pd.set_option('display.max_columns',None)  #expand the output display to  see more columns
df = pd.read_csv('/content/drive/MyDrive/dataset/GB_Rl.csv')
df.isnull().sum()
```

```python
df.describe()
```

```python
df_final = df.dropna()
df_final
```

```python
df_final=df_final.drop(['Passengerid','zero', 'zero.1', 'zero.2', 'zero.3', 'zero.4', 'zero.5', 'zero.6',  'zero.7',
       'zero.8', 'zero.9', 'zero.10', 'zero.11', 'zero.12', 'zero.13',
       'zero.14',  'zero.15', 'zero.16', 'zero.17',
       'zero.18', ],axis=1)
```

```python
df_final = df_final.rename(columns={'2urvived':'Target'})
df_final.head()
```

```python
X = df_final.drop('Target',axis=1)
y = df_final['Target']
print('shape of X:',X.shape)
print('shape of y:',y.shape)
```

```python
scaler = MinMaxScaler()
X_new = scaler.fit_transform(X)
```

```python
Xtrain,Xtest,ytrain,ytest = train_test_split(X_new,y,test_size=0.2,random_state=25)
print('shape of Xtrain:',Xtrain.shape)
print('shape of Xtest:',Xtest.shape)
print('shape of ytrain:',ytrain.shape)
print('shape of ytest:',ytest.shape)
```

```python
gb_clf = GradientBoostingClassifier(n_estimators=40,learning_rate=0.2,max_depth=2,max_features=11,random_state=45)
gb_clf.fit(Xtrain,ytrain)
y_pred = gb_clf.predict(Xtest)
accuracy_score(y_pred,ytest)
```
