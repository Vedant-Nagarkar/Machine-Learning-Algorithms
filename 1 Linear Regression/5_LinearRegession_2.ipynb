```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_boston
```

```python
#load the dataset
boston = load_boston()

#convert the dataset to a pandas dataframe
df = pd.DataFrame(boston.data, columns=boston.feature_names)
df['MEDV'] = boston.target
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df[boston.feature_names],
                                   df['MEDV'], test_size=0.1, random_state=1)

print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression

#create a linear regression object
lr = LinearRegression()

#train the model using the training sets
lr.fit(X_train,y_train)

#print the coefficient
print('coefficients:',lr.coef_)
print('intercept:',lr.intercept_)

```

```python
#Make predictions on the test set
y_pred = lr.predict(X_test)

#Compute the mean squared error(mse)
from sklearn.metrics import mean_squared_error

mse = mean_squared_error(y_test,y_pred)
print('MSE',mse)

#compute the root mean squared error (RMSE)
rmse = np.sqrt(mse)
print('RMSE',rmse)

#Compute the total sum of square(TSS)
tss = np.sum(y_test - np.mean(y_test)**2)
print('TSS',tss)

#Compute the sum of squares of residuals(SSR)
ssr = np.sum((y_test - y_pred)**2)
print('SSR',ssr)

#Compute the coefficient of determination(R^2)
r_squared = 1 - (ssr/tss)
print('R^2',r_squared)

#Compute the adjusted R^2
n = X_test.shape[0]
p = X_test.shape[1]
adjusted_r_squared = 1 - ((1-r_squared) * (n - 1)) / (n - p - 1 )
print('Adjusted R^2',adjusted_r_squared)
```
